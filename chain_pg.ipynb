```python
import gym
from gym import spaces
from gym.utils import seeding
from gym.envs.registration import register

class chain_pg():
    def __init__(self, chain_len=5):
        self.len_chain = chain_len #length of one chain
        self.n = self.len_chain*5 + 2 #length of MDP
        self.state = 0  # Start at beginning of the chain
        self.action_space = spaces.Discrete(2) #number of actions - 2 - [0: go left, 1: go right]
        self.pos_reward = 10
        self.neg_reward = -10
        self.observation_space = spaces.Discrete(self.n)
        self.seed()
        self.bottle_neck_1 = self.len_chain + 1
        self.bottle_neck_2 = (self.len_chain * 2) + 2

    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]

    def step(self, action):
        '''
        takes an action as an argument and returns the next_state, reward, done, info.
        '''
        assert self.action_space.contains(action)
        if self.state in [(self.len_chain * 3)+2, (self.len_chain * 4)+2]:
            reward = -10
            done = True
            self.state = 0
            return self.state, reward, done, {}
        
        elif self.state == (self.len_chain * 5)+2:
            reward = 10
            done = True
            self.state = 0
            return self.state, reward, done, {}
            
        elif self.state == self.bottle_neck_1:
            reward = 0
            done = False
            if not action:
                self.state += 1
            else:
                self.state = (self.len_chain * 3) + 3
            return self.state, reward, done, {}
            
        elif self.state == self.bottle_neck_2:
            reward = 10
            done = False
            if not action:
                self.state += 1
            else:
                self.state = (self.len_chain * 4) + 3
            return self.state, reward, done, {}
        
        else:
            reward = 0
            done = False
            if self.state == (self.len_chain*2)+1:
                reward = 10
            self.state += 1
            return self.state, reward, done, {}
        

    def reset(self):
        '''
        transitions back to first state
        '''
        self.state = 0
        return self.state
    
register(
    id='Chain_PG-v0',
    entry_point='chains:chain_pg',
    timestep_limit=20000,
    reward_threshold=1,
)
```
